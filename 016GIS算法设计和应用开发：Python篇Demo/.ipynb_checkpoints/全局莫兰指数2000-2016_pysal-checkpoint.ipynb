```python
sacField = ["GDP{0}".format(i) for i in range(2000,2017)]
```

```python
wqueen= pysal.queen_from_shapefile(shpfile)
```

```python
mv = []
for f in sacField:
    y = dbf.by_col(f)
    m = pysal.Moran(y,wqueen)
    mv.append({"_year":int(f.split("GDP")[1]),
              "moran'sI":float(m.I),
              "z_score":m.z_rand,
              "p_value":m.p_rand})
mpd = pd.DataFrame(mv)
```
