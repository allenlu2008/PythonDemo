### 定义个叫做狗的类，Python3支持中文变量名，但是实际开发中，最好不要这么做

```python
class 狗:
    名字 = ""
    品种 = ""
    年龄 = ""
    def __init__(self,名字,品种,年龄):
        self.名字 = 名字
        self.品种 = 品种
        self.年龄 = 年龄
    def 狗牌(self):
        print("名字 ： {0}".format(self.名字))
        print("品种 ： {0}".format(self.品种))
        print("年龄 ： {0}岁".format(self.年龄))
    def 打(self,次数=1):
        for i in range(次数):
            print("汪")
```

```python
甲 = 狗("托尼","泰迪",3)
```

```python
甲.狗牌()
```

```python
甲.打(10)
```

### 定义一个新的类，叫做傻狗，继承了狗的所有属性和方法，并且有自己特有的方法

```python
class 傻狗(狗):
    def 打(self,次数=1):
        if 次数 ==1:
            print("跑掉")
        else:
            print("我跑远了，你打不到我了")
```

```python
乙 = 傻狗("约翰","二哈",5)
```

```python
乙.狗牌()
```

```python
乙.打()
```

```python
乙.打(5)
```
